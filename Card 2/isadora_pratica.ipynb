```python
# ====================================================
# IMPORTEI A FERRAMENTA QUE ME DÁ ACESSO RÁPIDO A LLMs
# ====================================================
import os
from google.colab import userdata
from groq import Groq

client = Groq(
    api_key=userdata.get('fastcamp'),
)
```

```python
prompt = """
Você corre em um loop de Pensamento, Ação, PAUSA, Observação.
No final do loop, você emite uma Resposta
Use Pensamento para descrever seus pensamentos sobre a pergunta que lhe foi feita.
Use Ação para executar uma das ações disponíveis para você - então retorne PAUSA.
A Observação será o resultado da execução dessas ações.
Você deve utilizar SOMENTE informações do DB, sem fazer uso de conhecimento prévio.

Suas ações disponíveis são:

buscar:
por exemplo, buscar: "Daenerys"
Busca por essa personagem e retorna o nome da atriz -> Emilia Clarke

Sessão de exemplo:

Pergunta: Quem é a atriz que interpreta a Daenerys?
Pensamento: Eu deveria procurar Daenerys no DB de personagens disponível
Ação: Banco de Dados: Daenerys
PAUSA

Você será chamado novamente com isto:

Observação: Daenerys é uma personagem do seriado Game Of Thrones, inspirado na saga de livros escrita por George R. R. Martin. Essa personagem foi interpreta por Emilia Clarke.

Você então produz:

Resposta: Daenerys Targaryen foi interpretada por Emilia Clarke."""
```

```python
class AgentePersonagens:
    def __init__(self, client, system) -> None:
        self.client = client
        self.system = system
        self.messages = []

    if self.system:
        self.messages.append({
            "role": "system",
            "content": self.system,
    })

    def call(self, message=""):
        "Método que permite o usuário realizar uma consulta"
        if message:
        self.messages.append({
            "role": "user",
            "content": message,
        })
        result = self.execute() # chamamos o método execute para processar a iteração e ter uma resposta do modelo
        self.messages.append({
            "role": "assistant",
            "content": result,
        })
        return result

    def execute(self):
        completion = self.client.chat.completions.create(
            messages=self.messages, # recebe o histórico de mensagens
            model="llama-3.3-70b-versatile"
        )
        return completion.choices[0].message.content
```

```python
import re

def agent_loop(max_iterations, system, query):
  agent = AgentePersonagens(client, prompt)
  tools = ['db_personagens']
  next_prompt = query
  i = 0
  while i < max_iterations:
    i += 1
    result = agent.call(next_prompt)
    print(result)

    if "PAUSE" in result and "Action" in result:   # -> Analisando o resultado
      action = re.findall(r"Action: ([a-z]+): (.+)", result, re.IGNORECASE)
      chosen_tool = action[0][0]
      arg = action[0][1]

      if chosen_tool in tools:
        result_tool = eval(f"{chosen_tool}('{arg}')")
        next_prompt = f"Observation: {result_tool}"

      else:
        next_prompt = "Observation: Tool not found"

      print(next_prompt)
      continue

    if "Answer" in result:
      print(result)
      break
```

```python
# ====================================================
# IMPORTEI A FERRAMENTA QUE ME DÁ ACESSO RÁPIDO A LLMs
# ====================================================
import os
import json
from google.colab import userdata
from groq import Groq
import json
import requests
from bs4 import BeautifulSoup
import re
import urllib.parse
from typing import List, Dict, Any, Optional


client = Groq(
    api_key=userdata.get('fastcamp'),
)
```

```python
sites_confiaveis = [
    "petz.com.br",
    "cobasi.com.br",
    "petlove.com.br",
    "portaldodog.com.br",
    "royalcanin.com.br"
]
```

```python
class PesquisaWebTool:
    def __init__(self, sites_confiaveis):
        self.sites_confiaveis = sites_confiaveis
        self.name = "pesquisa_web"  # Nome que damos a nossa ferramenta
        self.description = "Busca informações em sites confiáveis sobre tópicos relacionados a pets"
        self.parameters = {            # Define os parâmetros que a nossa ferramenta aceita
            "type": "object",
            "properties": {
                "query": {
                    "type": "string",
                    "description": "Consulta de busca específica para encontrar informações sobre pets"
                },
                "site": {
                    "type": "string",
                    "description": "Site específico para buscar (opcional)",
                    "enum": sites_confiaveis if sites_confiaveis else ["todos"]
                }
            },
            "required": ["query"]
        }

    def execute(self, query, site=None):
        """Executa a busca na web com a query fornecida"""
        resultados = []

        if site and site in self.sites_confiaveis:
            # Busca apenas no site especificado pelo usuário
            site_resultados = self._pesquisa_site(site, query)
            resultados.extend(site_resultados)
        else:
            # Busca em todos os sites confiáveis
            for site in self.sites_confiaveis:
                site_resultados = self._pesquisa_site(site, query)
                resultados.extend(site_resultados)

        return resultados

    def _pesquisa_site(self, site, query):
        """Método para buscar em um site específico"""
        try:
            if "petz.com.br" in site:
                search_url = f"https://www.petz.com.br/busca?q={urllib.parse.quote(query)}"
            elif "cobasi.com.br" in site:
                search_url = f"https://www.cobasi.com.br/busca?q={urllib.parse.quote(query)}"
            elif "petlove.com.br" in site:
                search_url = f"https://www.petlove.com.br/busca?q={urllib.parse.quote(query)}"
            elif "cachorro-verde.com" in site:
                search_url = f"https://cachorro-verde.com/?s={urllib.parse.quote(query)}"
            elif "portaldodog.com.br" in site:
                search_url = f"https://portaldodog.com.br/?s={urllib.parse.quote(query)}"
            else:
                # URL genérica de pesquisa para outros sites
                search_url = f"https://{site}/search?q={urllib.parse.quote(query)}"

            headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
            }

            # requisição HTTP com timeout de 10 segundos
            response = requests.get(search_url, headers=headers, timeout=10)

            if response.status_code == 200:
                soup = BeautifulSoup(response.text, 'html.parser') # analisa o HTML da página de resultados

                resultados = []

                # Lógica específica para o site da Petz
                if "petz.com.br" in site:
                    # Busca por divs que contêm produtos ou artigos
                    artigos = soup.find_all('div', class_=re.compile('product-item|article-item'))
                    for artigo in artigos[:5]:
                        titulo_elem = artigo.find('h2') or artigo.find('h3')
                        link_elem = artigo.find('a', href=True)
                        desc_elem = artigo.find('p', class_=re.compile('description|summary'))

                        if titulo_elem and link_elem:
                            titulo = titulo_elem.text.strip()
                            link = link_elem['href']
                            # Completa URLs relativas
                            if not link.startswith('http'):
                                link = f"https://www.petz.com.br{link}"
                            descricao = desc_elem.text.strip() if desc_elem else "Sem descrição disponível"

                            resultados.append({
                                "titulo": titulo,
                                "link": link,
                                "descricao": descricao,
                                "fonte": "Petz"
                            })

                elif "cobasi.com.br" in site:
                    produtos = soup.find_all('div', class_=re.compile('product-card|vtex-product-summary|product-summary'))
                    for produto in produtos[:5]:
                        titulo_elem = produto.find(['h1', 'h2', 'h3'], class_=re.compile('product-name|product-title'))
                        link_elem = produto.find('a', href=True)
                        preco_elem = produto.find('span', class_=re.compile('price|value'))

                        if titulo_elem and link_elem:
                            titulo = titulo_elem.text.strip()
                            link = link_elem['href']
                            if not link.startswith('http'):
                                link = f"https://www.cobasi.com.br{link}"

                            descricao = "Sem descrição disponível"
                            if preco_elem:
                                descricao = f"Preço: {preco_elem.text.strip()}"

                            resultados.append({
                                "titulo": titulo,
                                "link": link,
                                "descricao": descricao,
                                "fonte": "Cobasi"
                            })

                    # Busca também artigos do blog da Cobasi
                    artigos = soup.find_all('article') or soup.find_all('div', class_=re.compile('blog-post|article-card'))
                    for artigo in artigos[:3]:
                        titulo_elem = artigo.find(['h1', 'h2', 'h3'])
                        link_elem = artigo.find('a', href=True)
                        desc_elem = artigo.find('p', class_=re.compile('excerpt|summary|description'))

                        if titulo_elem and link_elem:
                            titulo = titulo_elem.text.strip()
                            link = link_elem['href']
                            if not link.startswith('http'):
                                link = f"https://www.cobasi.com.br{link}"
                            descricao = desc_elem.text.strip() if desc_elem else "Artigo do blog da Cobasi"

                            resultados.append({
                                "titulo": titulo,
                                "link": link,
                                "descricao": descricao,
                                "fonte": "Blog Cobasi"
                            })

                elif "petlove.com.br" in site:
                    produtos = soup.find_all('div', class_=re.compile('product-card|product-item|product-box'))
                    for produto in produtos[:5]:
                        titulo_elem = produto.find(['h1', 'h2', 'h3', 'h4']) or produto.find('span', class_=re.compile('name|title'))
                        link_elem = produto.find('a', href=True)
                        preco_elem = produto.find(['span', 'div'], class_=re.compile('price|value'))
                        brand_elem = produto.find(['span', 'div'], class_=re.compile('brand'))

                        if titulo_elem and link_elem:
                            titulo = titulo_elem.text.strip()
                            link = link_elem['href']
                            if not link.startswith('http'):
                                link = f"https://www.petlove.com.br{link}"

                            # Constrói uma descrição com mais informações se for possível
                            descricao_parts = []
                            if brand_elem:
                                descricao_parts.append(f"Marca: {brand_elem.text.strip()}")
                            if preco_elem:
                                descricao_parts.append(f"Preço: {preco_elem.text.strip()}")

                            descricao = " | ".join(descricao_parts) if descricao_parts else "Produto disponível na Petlove"

                            resultados.append({
                                "titulo": titulo,
                                "link": link,
                                "descricao": descricao,
                                "fonte": "Petlove"
                            })

                    artigos = soup.find_all('article') or soup.find_all('div', class_=re.compile('blog-post|article-item'))
                    for artigo in artigos[:3]:
                        titulo_elem = artigo.find(['h1', 'h2', 'h3'])
                        link_elem = artigo.find('a', href=True)
                        desc_elem = artigo.find('p') or artigo.find('div', class_=re.compile('excerpt|summary'))

                        if titulo_elem and link_elem:
                            titulo = titulo_elem.text.strip()
                            link = link_elem['href']
                            if not link.startswith('http'):
                                link = f"https://www.petlove.com.br{link}"
                            descricao = desc_elem.text.strip() if desc_elem else "Artigo do blog da Petlove"

                            resultados.append({
                                "titulo": titulo,
                                "link": link,
                                "descricao": descricao,
                                "fonte": "Blog Petlove"
                            })

                # Solução genérica caso não haja implementação ou se a específica não funcionar
                if not resultados:
                    artigos = soup.find_all(['article', 'div'], class_=re.compile('item|card|product|post'))

                    if not artigos:
                        titulos = soup.find_all(['h1', 'h2', 'h3'], limit=5)
                        for titulo in titulos:
                            link_elem = titulo.find_parent('a') or titulo.find('a')

                            if link_elem and 'href' in link_elem.attrs:
                                link = link_elem['href']
                                if not link.startswith('http'):
                                    link = f"https://{site}{link if link.startswith('/') else '/' + link}"

                                parent = titulo.parent
                                desc_elem = parent.find('p') if parent else None
                                descricao = desc_elem.text.strip() if desc_elem else "Sem descrição disponível"

                                resultados.append({
                                    "titulo": titulo.text.strip(),
                                    "link": link,
                                    "descricao": descricao,
                                    "fonte": site
                                })

                    for artigo in artigos[:5]:
                        titulo_elem = artigo.find(['h1', 'h2', 'h3', 'h4'])
                        link_elem = artigo.find('a', href=True) or (titulo_elem.find('a', href=True) if titulo_elem else None)
                        desc_elem = artigo.find('p')

                        if titulo_elem:
                            titulo = titulo_elem.text.strip()
                            link = link_elem['href'] if link_elem else "#"
                            if link and not link.startswith('http') and not link == "#":
                                link = f"https://{site}{link if link.startswith('/') else '/' + link}"

                            descricao = desc_elem.text.strip() if desc_elem else "Sem descrição disponível"

                            resultados.append({
                                "titulo": titulo,
                                "link": link,
                                "descricao": descricao,
                                "fonte": site
                            })

                return resultados

            else:
                return [{"erro": f"Falha ao acessar {site}. Código de status: {response.status_code}"}]

        except Exception as e:
            return [{"erro": f"Erro ao buscar em {site}: {str(e)}"}]

    def get_tool_config(self):
        """Retorna a configuração da ferramenta para o formato esperado pelo LLM"""
        return {
            "type": "function",
            "function": {
                "name": self.name,
                "description": self.description,
                "parameters": self.parameters
            }
        }
```

```python
class PetAssistant:
    def __init__(self, client, system, max_iterations, tools=None):
        self.client = client
        self.system = system
        self.max_iterations = max_iterations
        self.tools = tools if tools else []
        self.pet_perfil = {}
        self.tutor_perfil = {}
        self.tool_configs = [tool.get_tool_config() for tool in self.tools]
        self.tool_map = {tool.name: tool for tool in self.tools} # Cria um mapeamento de nome da ferramenta para o objeto da ferramenta

    def coletar_pet_info(self, info):
        """Coleta e armazena informações sobre o pet"""
        self.pet_perfil.update(info)

    def coletar_tutor_info(self, info):
        """Coleta e armazena informações sobre o tutor"""
        self.tutor_perfil.update(info)

    def gera_recomendacao(self, user_query):
        """Gera recomendações baseadas no perfil e ferramentas disponíveis"""
        mensagens = [
            {"role": "system", "content": self.system},
            {"role": "user", "content": f"""
            Perfil do Pet: {json.dumps(self.pet_perfil, indent=2)}
            Perfil do Tutor: {json.dumps(self.tutor_perfil, indent=2)}

            Consulta do usuário: {user_query}
            """}
        ]

        current_iteration = 0

        while current_iteration < self.max_iterations:
            current_iteration += 1
            response = self.client.chat.completions.create(
                model="llama-3.3-70b-versatile",
                messages=mensagens,
                tools=self.tool_configs if self.tool_configs else None,
                tool_choice="auto"
            )

            message = response.choices[0].message
            mensagens.append(message)

            # se a resposta tiver chamadas de ferramentas
            # Verifique se  message possui o atributo tool_calls
            if hasattr(message, 'tool_calls') and message.tool_calls:
                for tool_call in message.tool_calls:
                    function_name = tool_call.function.name

                    # Caso a ferramenta exista no mapeamento
                    if function_name in self.tool_map:
                        tool = self.tool_map[function_name] # Acessando ferramenta mapeada

                        try:
                            args = json.loads(tool_call.function.arguments)

                            if function_name == "pesquisa_web":
                                query = args.get("query")
                                site = args.get("site", None)

                                result = tool.execute(query, site)

                                # Adiciona o resultado na conversa
                                mensagens.append({
                                    "role": "tool",
                                    "tool_call_id": tool_call.id,
                                    "name": function_name,
                                    "content": json.dumps(result, indent=2)
                                })

                        except json.JSONDecodeError as e:
                            # se for passado algo errado ou se a ferramenta não existir
                            mensagens.append({
                                "role": "tool",
                                "tool_call_id": tool_call.id,
                                "name": function_name,
                                "content": json.dumps({"erro": f"Argumentos inválidos: {str(e)}"}, indent=2)
                            })
                    else:
                        mensagens.append({
                            "role": "tool",
                            "tool_call_id": tool_call.id,
                            "name": function_name,
                            "content": json.dumps({"erro": "Ferramenta não disponível"}, indent=2)
                        })
            else:
                break

        if current_iteration >= self.max_iterations: # vai gerar a nossa resposta final
            mensagens.append({
                "role": "user",
                "content": "Você atingiu o limite de iterações. Por favor, forneça sua melhor resposta com as informações disponíveis até agora."
            })

            final_response = self.client.chat.completions.create(
                model="llama-3.3-70b-versatile",
                messages=mensagens
            )

            return final_response.choices[0].message.content
        else:
            return mensagens[-1].content
```

```python
def agente_loop(client, system_prompt, iterations, sites_confiaveis):
    pesquisa_tool = PesquisaWebTool(sites_confiaveis=sites_confiaveis)

    pet_assistant = PetAssistant(
        client=client,
        system=system_prompt,
        max_iterations=5,
        tools=[pesquisa_tool]
    )

    print(10 * '=')
    print('CADASTRO')
    print(10 * '=')

    nome = input("Digite o nome do seu pet: ")
    especie = input("Digite a espécie do seu pet: ")
    raca = input("Digite a raça do seu pet: ")
    idade = input("Digite a idade do seu pet: ")
    peso = input("Digite o peso do seu pet (Kg): ")
    condicoes_saude = input("Digite as condições de saúde do seu pet (saudável ou condições separadas por vírgula): ").split(',')
    condicoes_saude = [condicao.strip() for condicao in condicoes_saude if condicao.strip()]
    estilo_de_vida = input("Digite o estilo de vida do seu pet (atividade física, alimentação, etc., separados por vírgula): ").split(',')
    estilo_de_vida = [pref.strip() for pref in estilo_de_vida if pref.strip()]

    pet_assistant.coletar_pet_info({
        "nome": nome,
        "especie": especie,
        "raca": raca,
        "idade": idade,
        "peso": peso,
        "condicoes_saude": condicoes_saude,
        "estilo_de_vida": estilo_de_vida
    })

    nome_tutor = input("\nDigite o seu nome: ")
    localizacao = input("Digite a sua localização: ")
    preferencias = input("Digite suas preferencias (ex: produtos naturais, serviços próximos): ").split(',')
    preferencias = [pref.strip() for pref in preferencias if pref.strip()]

    pet_assistant.coletar_tutor_info({
        "nome": nome_tutor,
        "localizacao": localizacao,
        "preferencias": preferencias
    })

    iteracao = 0

    while iteracao < iterations:
        iteracao += 1
        print(f'============ Iteração {iteracao} ============')
        prompt_atual = input("\nDigite sua consulta: ")

        resultado = pet_assistant.gera_recomendacao(prompt_atual)
        print("Assistente:", resultado)

        # print("Nenhuma ação ou resposta detectada. Finalizando.")
        # break

    if iteracao >= iterations:
        print(f"Atingido número máximo de iterações ({iterations}). Finalizando.")

    return "Consulta finalizada."
```
